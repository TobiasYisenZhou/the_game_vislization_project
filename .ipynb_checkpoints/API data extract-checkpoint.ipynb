```python
import os
from dotenv import load_dotenv
import requests
from requests.exceptions import HTTPError
import json
import pandas as pd

load_dotenv()
a=os.getenv("rawg")
b=os.getenv("b")
```

## End point: list of game

```python
url = f"https://rawg-video-games-database.p.rapidapi.com/games?key={a}"

headers = {
    'x-rapidapi-host': "rawg-video-games-database.p.rapidapi.com",
    'x-rapidapi-key': b
    }
try:
    response = requests.request("GET", url, headers=headers)
    response.raise_for_status()
except HTTPError as http_err:
    print(f'HTTP error occurred: {http_err}')
except Exception as err:
    print(f'Other error occurred: {err}')
else:
    r=json.dumps(response.json(), indent=2, sort_keys=True)
    print('Success!')
```

```python
json_f=response.json()
```

```python
json_f.keys()
```

```python
print(r)
```

```python
# top 20 games' detail
json_f["results"]
```

```python
# release number of each year
json_f["filters"]["years"]
```
